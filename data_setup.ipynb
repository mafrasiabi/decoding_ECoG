```python
import pandas as pd
import csv
import numpy as np
from scipy.io import loadmat
```

```python
hg = loadmat('HG.mat')
hg = hg["HG"]
erp = loadmat('ERP.mat')
erp = erp["ERP"] 

hg_pd = pd.DataFrame(pd.concat([pd.DataFrame(hg[0,0,i].T) for i in range(7)]))
hg_pr = pd.DataFrame(pd.concat([pd.DataFrame(hg[0,1,i].T) for i in range(7)]))

hg_wd = pd.DataFrame(pd.concat([pd.DataFrame(hg[1,0,i].T) for i in range(7)]))
hg_wr = pd.DataFrame(pd.concat([pd.DataFrame(hg[1,1,i].T) for i in range(7)]))

erp_pd = pd.DataFrame(pd.concat([pd.DataFrame(erp[0,0,i].T) for i in range(7)]))
erp_pr = pd.DataFrame(pd.concat([pd.DataFrame(erp[0,1,i].T) for i in range(7)]))

erp_wd = pd.DataFrame(pd.concat([pd.DataFrame(erp[1,0,i].T) for i in range(7)]))
erp_wr = pd.DataFrame(pd.concat([pd.DataFrame(erp[1,1,i].T) for i in range(7)]))
```

```python
frames1 = [hg_pd,hg_pr,hg_wd,hg_wr]
frames2 = [erp_pd,erp_pr,erp_wd,erp_wr]
tags1 = ['hpd', 'hpr', 'hwd', 'hwr']
tags2 = ['epd','epr','ewd','ewr']

y_crud1 =([np.repeat(tags1[i],frames1[i].shape[0],axis = 0).T for i in range(4)])
y_crud2 =([np.repeat(tags2[i],frames2[i].shape[0],axis = 0).T for i in range(4)])
y1 = [item for sublist in y_crud1 for item in sublist]
y2 = [item for sublist in y_crud2 for item in sublist]

x_df1 = pd.concat(frames1)
x_df2 = pd.concat(frames2)
x_df1['label'] = y1
x_df2['label'] = y2
```
